```python
import json 
```

```python
json1_file = open('all_characters.json')
json1_str = json1_file.read()
json1_data = json.loads(json1_str)
```

```python
json1_data
```

```python
all_char_list = []
for each in json1_data:
    all_char_list.append(json1_data[each]['name'].title() + ' ('+ json1_data[each]['series'].title() + ')')
```

```python
print(all_char_list)
```
