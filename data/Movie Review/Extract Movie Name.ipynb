```python
# pos train
pos_train_url = pd.read_csv('/Users/imac/Desktop/Movie Review/train/urls_pos.csv', header=None).rename(index=str, columns={0: 'link'})
pos_train_url['category'] = 'pos_train'
pos_train_url['review_content'] = ''
pos_train_url['rating'] = ''
```

```python
pos_train_all_files_path = "/Users/imac/Desktop/Movie Review/train/pos/"
pos_train_all_files_name_list = os.listdir(pos_train_all_files_path)
pos_train_all_files_name_list_complete = []
for each in pos_train_all_files_name_list:
    pos_train_all_files_name_list_complete.append(pos_train_all_files_path+each)
```

```python
pos_train_index_loc_list = []
pos_train_rating_list = []
```

```python
for each in pos_train_all_files_name_list:
    pos_train_index_loc_list.append(each.split('_')[0])
    pos_train_rating_list.append((each.split('_')[1]).split('.')[0])
```

```python
def read_first_line(file):
    with open(file, 'rt') as fd:
        first_line = fd.readline()
    return first_line
```

```python
pos_train_output_strings = []
for each in pos_train_all_files_name_list_complete:
    pos_train_output_strings.append(read_first_line(each))
```

```python
for i in range(len(pos_train_index_loc_list)):
    pos_train_url.at[str(pos_train_index_loc_list[i]), 'review_content'] = pos_train_output_strings[i]
    pos_train_url.at[str(pos_train_index_loc_list[i]), 'rating'] = pos_train_rating_list[i]
```
